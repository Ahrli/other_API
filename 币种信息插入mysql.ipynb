```python
from retry import retry

@retry(max_delay=3)
def get_url(url):
    response = requests.get(url=url,timeout=30)
    if response.status_code == 200:
        return True
    else:
        return False


lt = []
for i in data:
    dc = {'type': 1, 'exchange': 'okex', 'category': 'OKEX'}
    dc['code'] = 'OKEX.' + i['symbol']
    dc['name'] = i['name']
    dc['symbol'] = i['symbol']
    dc['img_url'] = None
    for token in i['tokens']:
        url = f'https://blockchains.oss-cn-shanghai.aliyuncs.com/static/coinInfo/{token}.png'
        if get_url(url):
            print(url)
            dc['img_url'] = url
            break
    lt.append(dc)
lt
```
